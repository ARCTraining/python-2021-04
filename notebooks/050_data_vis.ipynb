```python
# importing libraries for today
import os
import pandas as pd 
import plotnine as p9

%matplotlib inline 
```

```python
# filepath variable for where our data lives

filepath = "/content/gdrive/MyDrive/Colab Notebooks/intro-python-2021-04/data"
```

```python
# os.path.join is a nice way for handling file path concatenation
os.path.join(filepath,"surveys.csv")
```

```python
surveys_complete = pd.read_csv(os.path.join(filepath,"surveys.csv"))
```

```python
surveys_complete.head()
```

```python
surveys_complete.columns
```

```python
# create a copy of surveys_complete dataframe to
# quick demonstrate dropping NaNs from a subset of columns
survey_df = surveys_complete.copy()
```

```python
# drop NaNs from the subset of specified columns 

survey_df.dropna(subset=['record_id', 'month', 'day', 'year', 'plot_id', 'species_id', 'sex',
       'hindfoot_length'])
```

```python
# we need drop NaN values to avoid errors when plotting
surveys_complete.dropna(inplace=True)
```

```python
surveys_complete.head()
```

```python
# our first plotnine plot

(p9.ggplot(data = surveys_complete))
```

```python
(p9.ggplot(data = surveys_complete, 
           mapping = p9.aes(x = 'weight', y = 'hindfoot_length')))
```

```python
(p9.ggplot(data = surveys_complete, 
           mapping = p9.aes(x = 'weight', y = 'hindfoot_length'))
  + p9.geom_point()
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length'))
+ p9.geom_point()
)
```

```python
# creating a bar plot

(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'plot_id'))
+ p9.geom_bar()
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length'))
+ p9.geom_point(alpha = 0.1)
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length'))
+ p9.geom_point(alpha = 0.1, color = 'red')
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length', color = 'species_id'))
+ p9.geom_point(alpha = 0.5)
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length', color = 'species_id'))
+ p9.geom_point(alpha = 0.5)
+ p9.xlab("Weight (g)")
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length', color = 'species_id'))
+ p9.geom_point(alpha = 0.5)
+ p9.xlab("Weight (g)")
+ p9.scale_x_log10()
)
```

```python
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length', color = 'species_id'))
+ p9.geom_point(alpha = 0.5)
+ p9.xlab("Weight (g)")
+ p9.scale_x_log10()
+ p9.theme_538()
+ p9.theme(text = p9.element_text(size = 16))
)
```

```python
# plot species_id v weight

(p9.ggplot(data=surveys_complete,
           mapping = p9.aes(x = 'species_id', y = 'weight'))
  + p9.geom_point())
```

```python
(p9.ggplot(data=surveys_complete,
           mapping = p9.aes(x = 'species_id', y = 'weight'))
  + p9.geom_boxplot())
```

```python
# you can combine multiple geometry layers

(p9.ggplot(data=surveys_complete,
           mapping = p9.aes(x = 'species_id', y = 'weight'))
  + p9.geom_jitter(alpha = 0.2)
  + p9.geom_boxplot(alpha = 0.1))
```

```python
# we can use geom_smooth to add a layer that fits/smooths the data
# in this instance we fit a linear model 
(p9.ggplot(data = surveys_complete,
mapping = p9.aes(x = 'weight', y = 'hindfoot_length'))
+ p9.geom_point(alpha = 0.1)
+ p9.geom_smooth(method = 'lm', color='blue')
)
```

```python
surveys_complete.head()
```

```python
yearly_counts = surveys_complete.groupby(['year','species_id'])['species_id'].count()

yearly_counts = yearly_counts.reset_index(name = 'counts')
```

```python
yearly_counts
```

```python
(p9.ggplot( data = yearly_counts,
           mapping = p9.aes(x = 'year', y = 'counts', color = 'species_id'))
+ p9.geom_line()
)
```

```python
# create a plot object
my_plot = (p9.ggplot( data = yearly_counts,
           mapping = p9.aes(x = 'year', y = 'counts', color = 'species_id'))
+ p9.geom_line()
)
```

```python
my_plot
```

```python
# save the plot object with .save

my_plot.save(os.path.join(filepath, 'species_counts_overtime.png'),
             width = 10, height = 10, dpi = 300)
```
