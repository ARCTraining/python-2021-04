# Plotting with plotnine


Most tutorials will teach you about `matplotlib`. This is great for numerically focussed plotting, but if our work is more data-centric then there are more appropriate alternatives.

We like [`plotnine`](https://plotnine.readthedocs.io/en/stable) which is familiar to anyone who has used `ggplot2` in R.

```python
# And let's load our surveys data again (without the NA values)
surveys_complete = pd.read_csv('data/surveys.csv')
surveys_complete = surveys_complete.dropna()
```

```python
# Create the plot
surveys_plot = p9.ggplot (data = surveys_complete,
                          mapping = p9.aes(x = 'weight', y = 'hindfoot_length')) + p9.geom_point()


# Draw the plot
surveys_plot
```

## The process of building plots iteratively

```python
# We want to explore the desnity of the dense regions
# We could add transparency to avoid the overplotting
# Create the plot
surveys_plot = p9.ggplot (data = surveys_complete,
                          mapping = p9.aes(x = 'weight', y = 'hindfoot_length')) + p9.geom_point(alpha=0.05, color ='blue')


# Draw the plot
surveys_plot
```

```python
# Colour each species differently to show where they are in each cluster group
# Create the plot
surveys_plot = p9.ggplot (data = surveys_complete,
                          mapping = p9.aes(x = 'weight', 
                                           y = 'hindfoot_length',
                                           color = 'species_id')) + p9.geom_point(alpha=0.05)


# Draw the plot
surveys_plot
```

```python
# Add labels
# Create the plot
surveys_plot = p9.ggplot (data = surveys_complete,
                          mapping = p9.aes(x = 'weight', 
                                           y = 'hindfoot_length',
                                           color = 'species_id')) + p9.geom_point(alpha=0.05) + p9.xlab ("Weight (g)")


# Draw the plot
surveys_plot
```

```python
# Use a log axis to look in more detail at what is happening at the lower end
# Create the plot
surveys_plot = p9.ggplot (data = surveys_complete,
                          mapping = p9.aes(x = 'weight', 
                                           y = 'hindfoot_length',
                                           color = 'species_id')) \
+ p9.geom_point(alpha=0.1) \
+ p9.xlab ("Weight (g)") \
+ p9.scale_x_log10()


# Draw the plot
surveys_plot
```

```python
# Change the theme to make the background white and a font size of 16 for the text
# Create the plot
surveys_plot = p9.ggplot (data = surveys_complete,
                          mapping = p9.aes(x = 'weight', 
                                           y = 'hindfoot_length',
                                           color = 'species_id')) \
+ p9.geom_point(alpha=0.1) \
+ p9.xlab ("Weight (g)") \
+ p9.scale_x_log10() \
+ p9.theme_bw () \
+ p9.theme(text=p9.element_text(size = 16))



# Draw the plot
surveys_plot
```

```python
# Challenge 2
# Adapt the bar plot of the previous exercise by mapping the sex variable 
# to the color **fill** of the bar chart. Change the scale of the color fill by providing the 
# colors blue and orange manually
# (see API reference to find the appropriate function).
```

```python
# Create the plot layers
surveys_plot = p9.ggplot(data=surveys_complete,
                         mapping=p9.aes(x='plot_id',
                                        fill='sex')) \
+ p9.geom_bar() \
+ p9.scale_fill_manual(['blue', 'orange'])

# Plot
surveys_plot 
```

## Distributions

Visualising or plotting distributions is a common data analysis task

```python
# Let's say visualise the distribution of weight within each species id

# Create the plot layers
weight_distribution = p9.ggplot (data = surveys_complete,
                                 mapping = p9.aes(x = 'species_id',
                                                  y = 'weight')) \
+ p9.geom_boxplot()

# Plot
weight_distribution
```

```python
# We could overlay the individual obs to see the measurements and their distribution
# Create the plot layers
weight_distribution = p9.ggplot (data = surveys_complete,
                                 mapping = p9.aes(x = 'species_id',
                                                  y = 'weight')) \
+ p9.geom_jitter(alpha = 0.2) \
+ p9.geom_boxplot()

# Plot
weight_distribution
```

```python
# Challenge 3: Distributions
weight_distribution = p9.ggplot (data = surveys_complete,
                                 mapping = p9.aes(x = 'species_id',
                                                  y = 'weight',
                                                  color = 'factor(plot_id)')) \
+ p9.geom_jitter(alpha = 0.2) \
+ p9.geom_violin(alpha = 0, color ='0.7') \
+ p9.scale_y_log10()

# Plot
weight_distribution
```
