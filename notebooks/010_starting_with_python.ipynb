## User defined functions

```python
def add_function(a, b):
  result = a + b 
  return result
```

```python
add_function(10, 5)
```

```python
add_function("hello", "world")
```

```python
class newList(list):
  def add_function(self, key):
    return super(newList, self).__getitem__(key) \
    + super(newList, self).__getitem__(key + 1)
```

```python
val = [1, 2]

val = newList(val)

val.add_function(0)
```
