```python
# fibonacci sequence generating function
a= list(range(10))

for i in range(0, len(a)-2, +1):
  a[i+2] = a[i+1] + a[i]
```

```python
print(a)
```

```python
list(range(10))
```

```python
def fibbo_seq(a):

  fibbo_seq = [0, 1]

  if a == 0:

    print("Error")

  elif a == 1:

    print(fibbo_seq[a-1])

  elif a == 2:

    print(fibbo_seq)

  else:

    for i in range (2,a):

      next = fibbo_seq[i-1] + fibbo_seq[i-2]

      fibbo_seq.append(next)

    print(fibbo_seq)
```

```python
fibbo_seq(10)
```

```python
n = int(input("Enter your given length of a fibonacci sequence: "))

s1 = 0

s2 = 1

sum = 0

count = 1

while(count <= n):

  print(sum, end = " ")

  count += 1

  s1 = s2

  s2 = sum

  sum = s1 + s2
```

```python
def fib_rec(n):

  if n < 2:
    return n

  else:
    return fib_rec(n - 1) + fib_rec(n - 2)
```

```python
fib_rec(10)
```

```python
for x in range(10):
  print(fib_rec(x))
```

```python
[fib_rec(x) for x in range(10)]
```

```python
def fibs(n=10):
  a,b = 0, 1
  for x in range(n):
    print(a)
    a, b = b, a + b

fibs(10)
```
